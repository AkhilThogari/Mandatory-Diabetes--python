# **INTRODUCTION**

Mandaroty dataset diabetes. This dataset comes from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to  predict the diagnosis of whether  a patient has diabetes, based on certain diagnostic measures contained in the dataset.

**Variables used in the dataset**

* Pregnancies: Number of times pregnant

* Glucose: Plasma glucose concentration 2 hours in an oral glucose tolerance test

* Blood Pressure: Diastolic blood pressure (mm Hg)

* Skin Thickness: Triceps skin fold thickness (mm)

* Insulin: 2-Hour serum insulin (mu U/ml)

* BMI: Body mass index (weight in kg/(height in m)^2) 

* Diabetes Pedigree Function: Diabetes pedigree function

* Age: Age (years)

* Outcome: Class variable (0 or 1) 0 mean non-diabetic and 1 means diabetic

### **8.Please calculate the skewness value and divide variables into symmetrical, moderately skewed and highly skewed.**

```python
#Calculation the Skewness for df columns ,skeness formula is 3 * (Mean â Median) / Standard Deviation
def skewee_data(df):
  l ={}
  for i in df:
    if i == "Outcome" or i == "Age": #if the dataset countered outcome , age then it will not considered
      continue
    else:
      stds = 3 * (df[i].mean()-df[i].median()) /df[i].std() #calculating the skwness value
      l[i]=stds
  return l
```

```python
#Fairly Symmetrical means value should be b/w -0.5 to 0.5
#Moderate Skewed means value should be b/w 	-0.5 to -1.0 and 0.5 to 1.0
#Highly Skewed means value should be b/w 	< -1.0 and > 1.0
skew = skewee_data(df2)
for i in skew:
  if(skew[i]>-1 and skew[i]<-0.5 or skew[i]>0.5 and skew[i]<1):
    print(f"{i} and value {skew[i]} --- Moderately Skewed")
  elif(skew[i]<-1 or skew[i]>1):
    print(f"{i} and value {skew[i]} ----- High Skewed")
  elif(skew[i]>-0.5 and skew[i]<0.5):
    print(f"{i} and value {skew[i]} --- is Symmetrical")
    
```

**Observation:**
* The Pregnancies,Diabetes Pedigree Function Insulin are Highly Skewed.
* Glucose,BloodPressure,SkinThickness,BMI is Symmetrical , it is nor prefectly Symmetrical but it is Fairly Symmetrical.
